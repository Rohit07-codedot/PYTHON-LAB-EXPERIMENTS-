6) Hypotenuse using Pythagoras Theorem

Formula: c = √(a² + b²)

```python
a = 3
b = 4

c = (a*a + b*b) ** 0.5

print("First side:", a)
print("Second side:", b)
print("Hypotenuse:", c)
```

7) Simple Interest

Formula: SI = (P × R × T) / 100

```python
p = 1000
r = 5
t = 2

si = (p * r * t) / 100

print("Principal:", p)
print("Rate:", r)
print("Time:", t)
print("Simple Interest:", si)
```

8) Area of Triangle (Heron’s Formula)

Formula:
s = (a+b+c)/2
Area = √(s(s-a)(s-b)(s-c))

```python
a = 3
b = 4
c = 5

s = (a + b + c) / 2
area = (s * (s-a) * (s-b) * (s-c)) ** 0.5

print("Side 1:", a)
print("Side 2:", b)
print("Side 3:", c)
print("Area of triangle:", area)
```

9) Convert Seconds into Hours, Minutes, Seconds

```python
seconds = int(input("Enter total seconds:580 "))

hours = seconds // 3600
remaining = seconds % 3600
minutes = remaining // 60
secs = remaining % 60

print("Hours:", hours)
print("Minutes:", minutes)
print("Seconds:", secs)
```

10) Swap Two Numbers Without Extra Variable

11) Sum of First n Natural Numbers

```python
n = int(input("Enter n: "))

sum = n * (n + 1) // 2

print("Sum:", sum)
```

12) Truth Table for Bitwise Operators

```python
print("a b  a&b  a|b  a^b")
for a in [0,1]:
    for b in [0,1]:
        print(a, b, " ", a&b, "  ", a|b, "  ", a^b)
        
```

13) Left Shift and Right Shift

```python
num = 8

print("Original Number:", num)
print("Left Shift (num << 1):", num << 1)
print("Right Shift (num >> 1):", num >> 1)
```

14) Membership Operator (Number in Sequence)

```python
num = 56

sequence = (10, 20, 56, 78, 89)

print("Number:", num)
print("Is number in sequence?", num in sequence)
```

15) Membership Operator (Character in String)
